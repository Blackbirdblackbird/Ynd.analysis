```python
#–î–æ–±–∞–≤–∏–º –≤ —Ç–∞–±–ª–∏—Ü—É –Ω–æ–≤—ã–µ —Å—Ç–æ–ª–±—Ü—ã —Å–æ —Å–ª–µ–¥—É—é—â–∏–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏: —Ü–µ–Ω–∞ –æ–¥–Ω–æ–≥–æ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞, –¥–µ–Ω—å –Ω–µ–¥–µ–ª–∏ –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è, –º–µ—Å—è—Ü –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è
#–≥–æ–¥ –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è, —Ç–∏–ø —ç—Ç–∞–∂–∞ –∫–≤–∞—Ä—Ç–∏—Ä—ã, —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –¥–æ —Ü–µ–Ω—Ç—Ä–∞ –≥–æ—Ä–æ–¥–∞ –≤ –∫–∏–ª–æ–º–µ—Ç—Ä–∞—Ö

data['price_square_meter'] = data['last_price'] / data['total_area']#–ü–æ—Å—á–∏—Ç–∞–µ–º —Ü–µ–Ω—É –æ–¥–Ω–æ–≥–æ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞
data['price_square_meter'] = round(data['price_square_meter'], 2) #–¥–ª—è —É–¥–æ–±—Å—Ç–≤–∞ –æ—Å—Ç–∞–≤–∏–º 2 –∑–Ω–∞–∫–∞ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π
#print(data['price_square_meter'].head(10)) #–ø–æ—Å–º–æ—Ç—Ä–∏–º, —á—Ç–æ –ø–æ–ª—É—á–∏–ª–æ—Å—å

data['day_exposition'] = data['first_day_exposition'].dt.weekday #–¥–µ–Ω—å –Ω–µ–¥–µ–ª–∏ –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è
data['month_exposition'] = data['first_day_exposition'].dt.month #–º–µ—Å—è—Ü –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è
data['year_exposition'] = data['first_day_exposition'].dt.year #–≥–æ–¥ –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è

#data.head(10)#–ø–æ—Å–º–æ—Ç—Ä–∏–º, —á—Ç–æ –ø–æ–ª—É—á–∏–ª–æ—Å—å

def type_floor(row): #–Ω–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ —É–∫–∞–∂–µ–º —Ç–∏–ø —ç—Ç–∞–∂–∞ –∫–≤–∞—Ä—Ç–∏—Ä—ã
    floors_total = row['floors_total']
    floor = row['floor']
    if floor == 1:
        return '–ø–µ—Ä–≤—ã–π'
    elif floor == floors_total:
        return '–ø–æ—Å–ª–µ–¥–Ω–∏–π'
    elif 1 < floor < floors_total:
        return '–¥—Ä—É–≥–æ–π'
    
data['type_floor'] = data.apply(type_floor, axis = 1)


data['cityCenters_nearest_km'] = round(data['cityCenters_nearest'] / 1000) #–ø–µ—Ä–µ–≤–µ–ª–∏ –º–µ—Ç—Ä—ã –≤ –∫–º, –∏ –æ–∫—Ä—É–≥–ª–∏–ª–∏

data.head(10)
#print(data['type_floor'].unique())
```

```python
#–ò–∑—É—á–∏–º, —Å—Ä–µ–¥–Ω—é—é —Ü–µ–Ω—É –æ–¥–Ω–æ–≥–æ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞ –≤ 10 –Ω–∞—Å–µ–ª—ë–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–∞—Ö —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º —á–∏—Å–ª–æ–º –æ–±—ä—è–≤–ª–µ–Ω–∏–π.
cityname = data['locality_name'].value_counts().index.tolist()[:10]
price_square = [data[data['locality_name'] == city]['price_square_meter'].median().round(2) for city in cityname]
price_home = [data[data['locality_name'] == city]['last_price'].median() for city in cityname]
frame = {'–ì–æ—Ä–æ–¥': cityname, '–¶–µ–Ω–∞ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞': price_square, '–¶–µ–Ω–∞ –∫–≤–∞—Ä—Ç–∏—Ä—ã': price_home}
df = pd.DataFrame(frame).sort_values(by='–¶–µ–Ω–∞ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞', ascending=False)
df

#–°–∞–º–∞—è –¥–æ—Ä–æ–≥–∞—è –Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç—å –≤ –°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥–µ, —Å–∞–º–∞—è –¥–µ—à–µ–≤–∞—è –≤ –í—ã–±–æ—Ä–≥–µ. –û–±—É—Å–ª–æ–≤–ª–µ–Ω–æ —Ç–µ–º, —á—Ç–æ –í—ã–±–æ—Ä–≥ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ 140 –æ—Ç –æ–±–ª–∞—Å—Ç–Ω–æ–≥–æ —Ü–µ–Ω—Ç—Ä–∞.
```
